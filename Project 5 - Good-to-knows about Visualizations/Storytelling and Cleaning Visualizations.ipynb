```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
fig = plt.figure(figsize = (30, 20))
index = 0

for r in range(1, 19, 3):
    ax = fig.add_subplot(6, 3, r)
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[index]],
            c = cb_dark_blue, linewidth = 3, label = "Women")
    ax.plot(women_degrees["Year"], 100 - women_degrees[stem_cats[index]],
           c = cb_orange, linewidth = 3, label = "Men")
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    ax.set_title(stem_cats[index])
    ax.tick_params(bottom = "off", top = "off", left = "off", right = "off", 
                  labelbottom = "off")
    ax.set_yticks([0, 100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    if r == 1:
        ax.text(2005, 90, "Men")
        ax.text(2005, 10, "Women")
    elif r == 16:
        ax.text(2009, 35, "Women")
        ax.text(2009, 65, "Men")
        ax.tick_params(labelbottom = "on")
        
    if index != 5:
        ax2 = fig.add_subplot(6, 3, r + 1)
        ax2.plot(women_degrees["Year"], women_degrees[lib_arts_cats[index]], 
                 c = cb_dark_blue, linewidth = 3, label = "Women")
        ax2.plot(women_degrees["Year"], 100 - women_degrees[lib_arts_cats[index]],
                c = cb_orange, linewidth = 3, label = "Men")
        for key, spine in ax2.spines.items():
            spine.set_visible(False)
        ax2.set_xlim(1968, 2011)
        ax2.set_ylim(0, 100)
        ax2.set_title(lib_arts_cats[index])
        ax2.tick_params(bottom = "off", top = "off", left = "off", right = "off", 
                  labelbottom = "off")
        ax2.set_yticks([0, 100])
        ax2.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
        if r + 1 == 2:
            ax2.text(2009, 20, "Men")
            ax2.text(2009, 80, "Women")
        elif r + 1 == 14:
            ax2.tick_params(labelbottom = "on")
    
    ax3 = fig.add_subplot(6, 3, r + 2)
    ax3.plot(women_degrees["Year"], women_degrees[other_cats[index]],
            c = cb_dark_blue, linewidth = 3, label = "Women")
    ax3.plot(women_degrees["Year"], 100 - women_degrees[other_cats[index]],
           c = cb_orange, linewidth = 3, label = "Men")
    for key, spine in ax3.spines.items():
        spine.set_visible(False)
    ax3.set_xlim(1968, 2011)
    ax3.set_ylim(0, 100)
    ax3.set_title(other_cats[index])
    ax3.tick_params(bottom = "off", top = "off", left = "off", right = "off", 
                  labelbottom = "off")
    ax3.set_yticks([0, 100])
    ax3.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    if r + 2 == 3:
        ax3.text(2009, 15, "Men")
        ax3.text(2009, 85, "Women")
    elif r + 2 == 18:
        ax3.text(2009, 65, "Men")
        ax3.text(2009, 30, "Women")
        ax3.tick_params(labelbottom = "on")
    
    index += 1

plt.show()
```
